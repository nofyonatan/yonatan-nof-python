### Part 1: What is a Function?
A function is a reusable block of code that performs a specific task. Functions help organize code and avoid repetition.  
Syntax:

```python
def function_name():
    # code to run
    pass
```

Code cell:


```python
# Define a simple function
def greet():
    print("Hello!")
    print("Welcome to Python")

# Call the function
greet()
greet()  # Can call multiple times
```

--- 
### Part 2: Functions with Parameters
Parameters let you pass information into functions.  
Code cell:

```python
# Function with one parameter
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")

# Function with multiple parameters
def introduce(name, age):
    print(f"My name is {name} and I am {age} years old")

introduce("Charlie", 15)
introduce("Diana", 16)
```

---
### Part 3: Return Values
Functions can send back a result using return.  
Code cell:

```python
# Function that returns a value
def add(a, b):
    result = a + b
    return result

# Use the returned value
total = add(5, 3)
print(total)  # 8

print(add(10, 20))  # 30

# Function with calculation
def square(number):
    return number * number

print(square(5))   # 25
print(square(10))  # 100
```

---
### Part 4: Functions with Default Parameters
You can give parameters default values.  
Code cell:

```python
# Default parameter
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet("Alice")    # Hello, Alice!
greet()           # Hello, Guest!

# Multiple defaults
def calculate_price(price, tax=0.08):
    total = price + (price * tax)
    return total

print(calculate_price(100))       # 108.0 (uses default tax)
print(calculate_price(100, 0.10)) # 110.0 (custom tax)
```

---
### Part 5: Variable Scope
Variables created inside functions are local - they only exist inside that function.  
Code cell:

```python
# Local vs Global variables
total = 0  # Global variable

def add_numbers(a, b):
    result = a + b  # Local variable
    return result

answer = add_numbers(5, 3)
print(answer)  # 8
# print(result)  # This would cause an error - result doesn't exist here
```

---
### Part 6: Documenting Functions
Use docstrings (triple quotes) to document what a function does.  
Code cell:

```python
def calculate_area(width, height):
    """
    Calculate the area of a rectangle.
    
    Parameters:
    width - the width of the rectangle
    height - the height of the rectangle
    
    Returns:
    The area as width * height
    """
    return width * height

# Use the function
area = calculate_area(5, 10)
print(area)

# View documentation
print(calculate_area.__doc__)
```

---
### Important Notes

- Use def to define a function
- Function names should be lowercase with underscores
- Parameters go in parentheses
- Use return to send back a value
- Call functions by using their name with parentheses
- Functions make code reusable and organized

### Task 5: Grade Calculator
Create a function get_grade(score) that:

1. Takes a test score (0-100)
2. Returns the letter grade:

    - 90-100: "A"
    - 80-89: "B"
    - 70-79: "C"
    - 60-69: "D"
    - Below 60: "F"


3. Test with different scores

Code cell:

```python
# Your code here
def get_grade(score):
    if score >= 90 and score <= 100:
        return "A"
    elif score >= 80 and score <= 89:
        return "B"
    elif score >= 70 and score <= 79:
        return "C"
    elif score >= 60 and score <= 69:
        return "D"
    else:
        return "F"
    
grade = get_grade(93)
print(grade)
grade = get_grade(67)
print(grade)
grade = get_grade(43)
print(grade)
```

### Task 6: List Statistics (Challenge)
Create a function analyze_list(numbers) that:

1. Takes a list of numbers as parameter
2. Calculates and returns three values:

    - Sum of all numbers
    - Average
    - Count of numbers


3. Use the function with a list of test scores

Hint: A function can return multiple values: return sum, avg, count

```python
# Your code here
def analyze_list(numbers):
    sum_numbers = sum(numbers)
    count_numbers = len(numbers)
    average_numbers = sum_numbers/count_numbers
    return sum_numbers, count_numbers, average_numbers

sum, count, average = analyze_list([80, 95, 100])
print(sum)
print(count)
print(average)
```

Code cell for homework:

```python
# Space for your homework
def power(base, exponent):
    result =  base ** exponent
    return result

def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

answer = power(5, 2)
print(answer)
answer = factorial(4)
print(answer)
```

```python

def count_vowels(text):
    number_of_vowels = 0
    for char in text.lower():
        if "a" in char or "e" in char or "i" in char or "o" in char or "u" in char:
            number_of_vowels += 1
    return number_of_vowels

def reverse_string(text):
    reverse_text = text[::-1]
    return reverse_text

vowels = count_vowels("Are you okey?")
print(vowels)

reverse = reverse_string("hello")
print(reverse)
```

```python
def is_valid_email(email):
    if "@" in email and "." in email:
        return "valid"
    else:
        return "invalid"
    
def is_strong_password(password):
    password_length = len(password)
    if password_length >= 8:
        return "Strong password"
    else:
        return "Weak password"
    
email = is_valid_email("Dani@gameil.com")
print(email)
email = is_valid_email("Dani.dani")
print(email)

password = is_strong_password("12345678")
print(password)
password = is_strong_password("6732")
print(password)
```
